```python
import pandas as pd
import xgboost as xgb
import optuna
import numpy as np
import pickle

from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
from sklearn.preprocessing import MinMaxScaler, StandardScaler

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
train_path = '../data/clean/train.csv'
df = pd.read_csv(train_path, index_col='date')
df.head()
#df = df.loc[df.index >= '2022-01-01']

df.dtypes
```

**MAE**: Set objective to ***reg:absoluteerror*** and return mae from the objective function.\
**MAPE**: Set objective to ***reg:squarederror***, calculate MAPE manually, and return it from the objective function.\
**R²**: Set objective to ***reg:squarederror***, calculate R² manually, and return -r2 from the objective function to maximize R².\
**MSE**: Set objective to ***reg:squarederror*** and return mse from the objective function.\
**RMSE**: Set objective to ***reg:squarederror***, calculate RMSE manually, and return it from the objective function.\
**sMAPE**: Set objective to ***reg:squarederror***, calculate sMAPE manually, and return it from the objective function.

```python
def load_data(path):
    df = pd.read_csv(path, index_col='date')
    df.index = pd.to_datetime(df.index)
    
    #features = df[['open', 'high', 'low', 'close', 'volume', 'change', 'changePercent',
    #               'vwap', 'sma', 'ema', 'wma', 'dema', 'tema', 'williams', 'rsi', 'adx',
    #               'ratingScore', 'minus_7_price', 'minus_5_price', 'minus_3_price']]

    features = df[['open', 'high', 'low',
                   'vwap', 'dema', 'tema', 'williams', 'rsi',
                   'ratingScore', 'minus_7_price', 'minus_5_price', 'minus_3_price']]
    
    target = df[['target']]
    other = df.drop(columns=features.columns.tolist() + target.columns.tolist())
    
    return features, target, other


def normalize_data(train_features, test_features):
    normalizer = MinMaxScaler()
    normalizer.fit(train_features)

    filename = "normalizer.pkl"
    with open("../scalers/" + filename, "wb") as file:
        pickle.dump(normalizer, file)

    train_features_n = normalizer.transform(train_features)
    train_features_df = pd.DataFrame(train_features_n, columns=train_features.columns, index=train_features.index)

    test_features_n = normalizer.transform(test_features)
    test_features_df = pd.DataFrame(test_features_n, columns=test_features.columns, index=test_features.index)

    return train_features_df, test_features_df


def objective(trial):
    # Load your stock price dataset
    features, target, other = load_data('../data/clean/train.csv')
    
    # Split data into training and validation sets
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, shuffle=False)
    X_train, X_test = normalize_data(X_train, X_test)
    
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dvalid = xgb.DMatrix(X_test, label=y_test)
    
    param = {
        "verbosity": 0,
        "objective": "reg:squarederror",  # Regression objective (try "reg:linear" ???"reg:squarederror")
        "tree_method": "exact",
        "booster": trial.suggest_categorical("booster", ["gbtree", "gblinear", "dart"]),
        "lambda": trial.suggest_float("lambda", 1e-8, 1.0, log=True),
        "alpha": trial.suggest_float("alpha", 1e-8, 1.0, log=True),
        "subsample": trial.suggest_float("subsample", 0.2, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
        "n_estimators": trial.suggest_int("n_estimators", 50, 500)  # Adding n_estimators
    }
    
    if param["booster"] in ["gbtree", "dart"]:
        param["max_depth"] = trial.suggest_int("max_depth", 3, 9, step=2)
        param["min_child_weight"] = trial.suggest_int("min_child_weight", 2, 10)
        param["eta"] = trial.suggest_float("eta", 1e-8, 1.0, log=True) # learning_rate
        param["gamma"] = trial.suggest_float("gamma", 1e-8, 1.0, log=True)
        param["grow_policy"] = trial.suggest_categorical("grow_policy", ["depthwise", "lossguide"])
    
    if param["booster"] == "dart":
        param["sample_type"] = trial.suggest_categorical("sample_type", ["uniform", "weighted"])
        param["normalize_type"] = trial.suggest_categorical("normalize_type", ["tree", "forest"])
        param["rate_drop"] = trial.suggest_float("rate_drop", 1e-8, 1.0, log=True)
        param["skip_drop"] = trial.suggest_float("skip_drop", 1e-8, 1.0, log=True)
    
    bst = xgb.train(param, dtrain)
    preds = bst.predict(dvalid)
    rmse = mean_squared_error(y_test, preds)
    
    return rmse


# Suppress Optuna's logging output
optuna.logging.set_verbosity(optuna.logging.WARNING)


study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=300, timeout=600)
    
print("Number of finished trials: ", len(study.trials))
print("Best trial:")
trial = study.best_trial
    
print("  Value: {}".format(trial.value))
print("  Params: ")
for key, value in trial.params.items():
    print("    {}: {}".format(key, value))

```

```python
# THIS CELL NEED TO BE DUPLICATED AND TESTED WITH A DATASET THAT HAS THE SAME STRUCTURE AS THE TRAINING DATASET
# BUT INCLUDES THE LATEST DATA (PLUS ROWS FOR THE FUTURE SOMEHOW).
# IS THIS CELL


best_params = trial.params
path = '../data/clean/train.csv'

def prep_data(path):
    features, target, other = load_data(path)
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, shuffle=False)
    X_train, X_test = normalize_data(X_train, X_test)

    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = prep_data(path)

reg = xgb.XGBRegressor(early_stopping_rounds=50, **best_params)
reg.fit(X_train, y_train, eval_set=[(X_train, y_train),(X_test, y_test)], verbose=100)
#reg = xgb.XGBRegressor(**best_params)
#reg.fit(X_train, y_train)

pred = reg.predict(X_test)

print(f"MSE: {mean_squared_error(pred, y_test): .2f}")
print(f"MAE: {mean_absolute_error(pred, y_test): .2f}")
print(f"RMSE: {root_mean_squared_error(pred, y_test): .2f}")
print(f"R2 score:  {reg.score(X_test, y_test): .2f}")


# Get and print feature importances
bst = reg.get_booster()

```
